```python
import numpy as np

np.random.seed(42)

# Sales between 50 and 1000
df['Sales'] = np.random.randint(50, 1000, len(df))

# Quantity between 1 and 10
df['Quantity'] = np.random.randint(1, 10, len(df))

# Discount between 0% and 50%
df['Discount'] = np.round(np.random.uniform(0, 0.5, len(df)), 2)

# Profit between -100 and 300
df['Profit'] = np.round(np.random.uniform(-100, 300, len(df)), 2)

# Random Categories
df['Category'] = np.random.choice(
    ['Furniture', 'Office Supplies', 'Technology'],
    len(df)
)

# Random Sub-Categories
df['Sub-Category'] = np.random.choice(
    ['Chairs', 'Tables', 'Phones', 'Binders', 'Storage', 'Accessories'],
    len(df)
)

# Random Product Names
df['Product Name'] = np.random.choice(
    [
        'Office Chair',
        'Laptop',
        'Printer',
        'Bookcase',
        'Desk',
        'Monitor',
        'Keyboard',
        'Mouse',
        'Storage Box',
        'Phone'
    ],
    len(df)
)

# Random Regions
df['Region'] = np.random.choice(
    ['East', 'West', 'Central', 'South'],
    len(df)
)

# Random States
df['State'] = np.random.choice(
    ['California', 'Texas', 'New York', 'Florida', 'Washington'],
    len(df)
)

# Product IDs
df['Product ID'] = [
    f'P-{1000+i}' for i in range(len(df))
]

# Postal Codes
df['Postal Code'] = np.random.randint(10000, 99999, len(df))
```

```python
print(df.columns.tolist())
df.head()
```

```python
total_sales=df['Sales'].sum()

print(total_sales)
```

```python
total_profit = df['Profit'].sum()

print("Total Profit:", total_profit)
```

```python
profit_margin = (total_profit / total_sales) * 100

print("Profit Margin: {:.2f}%".format(profit_margin))
```

```python
total_orders = df['Order_ID'].nunique()

print("Total Orders:", total_orders)
```

```python
total_customers = df['Customer_ID'].nunique()

print("Total Customers:", total_customers)
```

```python
average_order_value = df.groupby('Order_ID')['Sales'].sum().mean()

print("Average Order Value:", round(average_order_value,2))
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

print(category_sales)
```

```python
import matplotlib.pyplot as plt

category_sales.plot(kind='bar', figsize=(8,5))

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
region_profit = df.groupby('Region')['Profit'].sum()

print(region_profit)
```

```python
region_profit.plot(kind='bar', color='orange', figsize=(8,5))

plt.title("Profit by Region")
plt.xlabel("Region")

plt.ylabel("Profit")
plt.show()
```

```python
monthly_sales = df.groupby('Month Name')['Sales'].sum()

print(monthly_sales)
```

```python
monthly_sales.plot(figsize=(12,5), marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

print(top_products)
```

```python
plt.figure(figsize=(10,6))

top_products.plot(kind='bar')

plt.title("Top 10 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Sales")

plt.xticks(rotation=45)
plt.show()
```

```python
worst_products = df.groupby('Product Name')['Profit'].sum().sort_values().head(10)

print(worst_products)
plt.figure(figsize=(10,6))

worst_products.plot(kind='barh', color='red')

plt.title("Bottom 10 Products by Profit")
plt.xlabel("Profit")
plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()
print(segment_sales)
plt.figure(figsize=(7,5))

segment_sales.plot(kind='pie', autopct='%1.1f%%')

plt.ylabel("")
plt.title("Sales by Segment")

plt.show()
```

```python
quantity = df.groupby('Category')['Quantity'].sum()

print(quantity)
plt.figure(figsize=(7,5))

quantity.plot(kind='bar')

plt.title("Quantity Sold by Category")
plt.xlabel("Category")
plt.ylabel("Quantity")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=df, x='Discount', y='Profit')

plt.title("Discount vs Profit")

plt.show()
```

```python
plt.figure(figsize=(8,6))

corr = df[['Sales', 'Profit', 'Quantity', 'Discount']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")

plt.show()
```
